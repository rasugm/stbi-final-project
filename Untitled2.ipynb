```python
import pandas as pd
import re
import json
from pymongo import MongoClient

# Define relevant keywords for each query (you need to replace these with actual keywords)
testing_queries = pd.read_csv('testing_queries.csv')

# Initialize an empty list to store dictionaries
relevance_keywords = []

# Iterate over the DataFrame rows
for index, row in testing_queries.iterrows():
    # Construct the dictionary for each row
    relevance_keyword = {'qid': row['qid'], 'vector': row['vector']}
    
    # Append the dictionary to the list
    relevance_keywords.append(relevance_keyword)

# Define documents (you need to replace these with actual documents)
client = MongoClient('mongodb://root:admin123%23@localhost:27017/?authMechanism=SCRAM-SHA-1&authSource=admin')
db = client['kpu']
collection_dataset_caleg_training = db['col_dataset_caleg_text']

# Retrieve data from MongoDB collection
data_collection = list(collection_dataset_caleg_training.find().limit(5))

# Convert ObjectId to string
for data in data_collection:
    data['_id'] = str(data['_id'])
    data['original_id'] = str(data['original_id'])

# Print the modified data_collection
documents = json.loads(json.dumps(data_collection))

# documents

# Create a DataFrame to store relevance judgments
testing_result = []

# Loop through each query and document pair
for relevance_keyword in relevance_keywords:
    qid = relevance_keyword['qid']
    vector = relevance_keyword['vector']
    keywords = vector.split()  # Split the vector into individual keywords
    for doc in documents:
        content = doc['text']
        # Check if the document contains any relevant keywords
        contains_keyword = any(re.search(r'\b{}\b'.format(keyword), content.lower()) for keyword in keywords)
        # Assign relevance judgment based on keyword matching
        rel = 1 if contains_keyword else 0
        # Append query, document, and relevance judgment to testing_result
        testing_result.append({'qid': qid, 'docid': doc['_id'], 'rel': rel})

# Convert testing_result to a DataFrame
testing_result = pd.DataFrame(testing_result)

testing_result.to_csv('testing_result.csv')
```
